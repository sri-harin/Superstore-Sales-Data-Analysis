```python
##GROUPBY
```

```python
#total sales by category
df.groupby("Category")["Sales"].sum()
```

```python
#total profits by category
df.groupby("Category")["Profit"].sum()
```

```python
#avg sales for each category
df.groupby("Category")["Sales"].mean()
```

```python
#max sales for each category
df.groupby("Category")["Sales"].max()
```

```python
#quantity sold in each region
df.groupby("Region")["Quantity"].sum()
```

```python
#DATA VISUALIZATION
#import library
import matplotlib.pyplot as plt
```

```python
#bar chart
df.groupby("Category")["Sales"].sum().plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
#line chart
df.groupby("Region")["Sales"].sum().plot(kind="line")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
#Pie chart
df.groupby("Category")["Sales"].sum().plot(kind="pie",autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.ylabel("")
plt.show()
```

```python
#Histogram
df["Sales"].plot(kind="hist")
plt.title("Sales Distribution")
plt.show()
```
